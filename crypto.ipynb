Checking Correlation between dependent and independent variables

1. Ripple

```python
ripple.head()
```

```python
sns.heatmap(ripple.corr(), cmap='coolwarm', annot=True);
```

```python
ripple.columns
```

```python
#Linear Model for Ripple

X = ripple[['Open', 'High', 'Low', 'Volume', 'Market.Cap']]
y = ripple['Close']
```

```python
#Training Set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
X_test.head()
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()

lm.fit(X_train, y_train)
```

```python
print(lm.intercept_)
```

```python
lm.coef_
```

```python
X_train.columns
```

```python
df_coeff = pd.DataFrame(lm.coef_, X.columns, columns=['Coeff'])

df_coeff
```

```python
#Testing Data

predictions = lm.predict(X_test)

predictions
```

```python
y_test.head()
```

```python
plt.scatter(y_test, predictions);
plt.xlabel('y-test')
plt.ylabel('predictions')
```

```python
sns.distplot((y_test - predictions));
```
